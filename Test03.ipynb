```python
#plotting a graph using matplotlib libraries.
import matplotlib.pyplot as plt

# Function to get user input for years and income
def get_input():
    years = []
    income = []
    print("Enter year and income. To stop, enter 'done'.")
    while True:
        year = input("Enter year (or 'done' to stop): ")
        if year.lower() == 'done':
            break
        income_val = input("Enter income: ")
        try:
            years.append(int(year))
            income.append(float(income_val))
        except ValueError:
            print("Please enter a valid year and income!")

    return years, income

# Get user input
years, income = get_input()

# Create the plot
plt.plot(years, income, marker='o')

# Set the title and labels
plt.title('Income over the Years')
plt.xlabel('Year')
plt.ylabel('Income')

# Display the plot
plt.grid(True)
plt.show()
```
